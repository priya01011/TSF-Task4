# Step 3 - Data Visualization



```python
print("Country with the most attacks:",df['Country'].value_counts().idxmax())
print("City with the most attacks:",df['City'].value_counts().index[1]) #as first entry is 'unknown'
print("Region with the most attacks:",df['Region'].value_counts().idxmax())
print("Year with the most attacks:",df['Year'].value_counts().idxmax())
print("Month with the most attacks:",df['Month'].value_counts().idxmax())
print("Group with the most attacks:",df['Group'].value_counts().index[1])
print("Most Attack Types:",df['Attack Type'].value_counts().idxmax())
```

# Yearwise Attacks

### Number of attacks in each year

```python
attacks=df["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")   
attacks.head()
```

```python
## Plotting the graph
attacks.plot(kind='bar',color='red',figsize=(15,12),fontsize=13)
plt.title('Timeline of Attacks in each year',fontsize=15)
plt.xlabel("Years",fontsize=12)
plt.legend(loc = 'upper left')
plt.ylabel('Numbers of attacks in each years',fontsize=15)
plt.show()
```

# Yearwise Casuality
### Number of casualities in each year

```python
yc=df[["Year","Casuality"]].groupby("Year").sum()
yc.head()
```

```python
## Plotting the graph

yc.plot(kind='bar',color='red',figsize=(15,10),fontsize=10 )
plt.title('Timeline of Casuality in each year',fontsize=12)
plt.xlabel("Years",fontsize=15,fontstyle='oblique')
plt.ylabel('Numbers of casualities each year',fontsize=15)
plt.legend(loc='upper left')
plt.show()
```

# Yearwise killed
## Number of killed in each year

```python
yk=df[["Year","Killed"]].groupby("Year").sum()
yk.head()
```

```python
# Plotting The graph
yk.plot(kind='bar',color='orange',figsize=(15,10),fontsize=15)
plt.title("Time-line of Killed people in each year",fontsize=15)
plt.xlabel('Years',fontsize=15,fontstyle='oblique')
plt.ylabel('Number of killed people in each year',fontstyle='oblique',fontsize=15)
plt.legend(loc='upper left')
plt.show()
```

# Region Wise Attacks

```python
reg=pd.crosstab(df.Year,df.region_txt)
reg.head()
```

```python
reg.shape
```

```python
## Plotting the graph
reg.plot(kind="area",stacked=False,alpha=0.5,figsize=(20,10))
plt.title("Region-wise attacks",fontsize=15)
plt.xlabel('Years',fontsize=15,fontstyle='oblique')
plt.ylabel('Number of attacks in each year',fontstyle='oblique',fontsize=15)
plt.legend(loc='upper left')
plt.show()
```

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(12,6),color = 'red')
plt.title("Total number of attacks from 1970-2017")
plt.xlabel('year')
plt.ylabel('Total number of attack')
plt.show()
```
